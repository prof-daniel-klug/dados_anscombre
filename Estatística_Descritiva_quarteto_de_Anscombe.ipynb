# **Visualização e medidas descritivas para um conjunto de dados**
O quarteto de Anscombe compreende quatro conjuntos de dados que têm *estatísticas descritivas* quase idênticas, mas têm distribuições muito diferentes quando analisadas em gráficos. Cada conjunto de dados consiste em onze pontos (x, y). Eles foram construídos em 1973 pelo estatístico Francis Anscombe para demonstrar tanto a importância do gráfico de dados antes de analisá-los quanto o efeito de outliers e outras observações influentes sobre propriedades estatísticas.
Fonte: https://en.wikipedia.org/wiki/Anscombe%27s_quartet

```python
# Instalação dos pacotes necessários
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Leitura do arquivo no meu Drive
# dados no git
url = 'https://raw.githubusercontent.com/prof-daniel-klug/dados_anscombre/main/dados_Anscombes.csvdados = pd.read_csv(url)
#dados = pd.read_csv('/content/drive/MyDrive/dados_Anscombes.csv')
```

```python
# visualisar os dados
dados.head(44)
```

```python
# Gráfico de dispersão, aspecto visual da Relação de X por Y - TODOS
fig  = px.scatter(dados, x = 'X', y = 'Y', width = 800)
fig.update_traces(marker = dict(size = 12, line=dict(width = 1)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Todos os grupos reunidos na relação X por Y')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
dados.groupby(by='GRUPO').describe()
```

```python
dados_A = dados[dados['GRUPO']=='A']
fig  = px.scatter(dados_A, x = 'X', y = 'Y', width = 800)
fig.update_traces(marker = dict(size = 12, line=dict(width = 2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Grupo A na relação X por Y')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
dados_B = dados[dados['GRUPO']=='B']
fig  = px.scatter(dados_B, x = 'X', y = 'Y', width = 800)
fig.update_traces(marker = dict(size = 12, line=dict(width = 2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Grupo B na relação X por Y')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
dados_C = dados[dados['GRUPO']=='C']
fig  = px.scatter(dados_C, x = 'X', y = 'Y', width = 800)
fig.update_traces(marker = dict(size = 12, line=dict(width = 2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Grupo C na relação X por Y')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
dados_D = dados[dados['GRUPO']=='D']
fig  = px.scatter(dados_D, x = 'X', y = 'Y', width = 800)
fig.update_traces(marker = dict(size = 12, line=dict(width = 2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Grupo D na relação X por Y')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
# Gráfico de dispersão com configurações de aspecto visual (Relação Geral X por Y)
fig  = px.scatter(dados, x = 'X', y = 'Y', width = 800, color='GRUPO')
fig.update_traces(marker = dict(size = 12, line=dict(width = 2)), selector = dict(mode = 'markers'))
fig.update_layout(title = 'Todos os grupos reunidos na relação X por Y com legenda de cores')
fig.update_xaxes(title = 'Valores do X')
fig.update_yaxes(title = 'Valores do Y')
fig.show()
```

```python
# Os quatro gráficos em sepadado
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(10,10))
ax1.scatter(x=dados_A['X'],y=dados_A['Y'],c='Blue',marker='o')
ax2.scatter(x=dados_B['X'],y=dados_B['Y'],c='Red',marker='o')
ax3.scatter(x=dados_C['X'],y=dados_C['Y'],c='Green',marker='o')
ax4.scatter(x=dados_D['X'],y=dados_D['Y'],c='Purple',marker='o')
plt.show()
```
